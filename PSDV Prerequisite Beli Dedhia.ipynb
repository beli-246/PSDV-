### 1. List Operations and Basic Statistics
- Create a list of the first 20 natural numbers.
- Compute the sum of all even numbers in the list.
- Compute the product of all odd numbers in the list.
- Calculate the mean, median, and standard deviation of the list.

```python
l=list(range(1,21))
s=0
p=1
for i in l:
    if i%2==0:
        s=s+i
    else:
        p=p*i

meanl=sum(l)/len(l)

def medianl(l):
    l.sort()
    k=len(l)
    if len(l)%2==0:
        m=l[(k//2)]+l[(k//2)+1]
    else:
        m=l[(k+1)//2]
    return m

for i in l:
    sv=0
    d=(i-meanl)**2
    sv=sv+d
    sd=(sv/len(l))**0.5

print(l)
print(s)
print(p)
print(meanl, medianl(l), sd)
```

### 3. Fibonacci Sequence:
- Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
- Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt
n=int(input())
def fibo(n):
    l = [0, 1]
    
    for i in range(2, n):
        k = l[-1] + l[-2]
        l.append(k)
    
    return l

print(fibo(n))
f = fibo(n)
plt.plot(range(1, n + 1), f)
plt.title('Fibonacci Sequence')
plt.xlabel('Index (n)')
plt.ylabel('Fibonacci Number')
plt.show()
```

### 13. Sorting Dictionaries:
- Write a function to sort a list of dictionaries by a key.
- Example:

- data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}] Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]



```python
data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}] 
for i in range(len(data)-1):
    if data[i]['age']>data[i+1]['age']:
        t=data[i]
        data[i]=data[i+1]
        data[i+1]=t
    else:
        continue

print(data)
```

- Visualize the word frequencies using a bar chart.


```python
import matplotlib.pyplot as plt
f = open("words.txt")
counts = dict()
for line in f:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word,0) + 1
maxcount = None
maxword = None
for k,v in counts.items():
    if maxcount is None or v > maxcount:
        maxword = k
        maxcount = v
        
print(maxword)

plt.bar(counts.keys(), counts.values())
plt.show()
```

### 21. Basic Statistics:
Write Python code to calculate:

- Mean, median, mode, and standard deviation of a list of numbers.
- Visualize the data using a box plot.

```python
import matplotlib.pyplot as plt
import math
import random
l = []
for i in range(100):
    l.append(random.randint(1,100))
mean = sum(l) / len(l)
l.sort()
median = 0
if len(l) % 2 == 0: 
    median1 = l[len(l)//2] 
    median2 = l[len(l)//2 - 1] 
    median = (median1 + median2)/2
else: 
    median = l[len(l)//2] 

f= {}
for num in l:
    if num in f:
        f[num] += 1
    else:
        f[num] = 1
mode = [num for num,freq in f.items() if freq == max(f.values())]
sd = math.sqrt(sum(pow(num-mean,2) for num in l)/len(l))
print(mean, median, mode, sd)
```
